valores al usuario.

```python
#Método dividir, solicita los nros al usuario y hace la división,
#si el nro divisor es 0 manda mensaje de que no debe ser cero
def dividir(dividendo, divisor):
  try:
    resultado = dividendo / divisor
    return print(f"El resultado de la división es: {resultado}")
  except ZeroDivisionError:
    print("El divisor no puede ser cero")
    #raise ZeroDivisionError("El divisor no puede ser cero")

#
ingresar_dividendo = int(input("Ingrese el dividendo "))
ingresar_divisor = int(input("Ingrese el divisor "))
resultado = dividir(ingresar_dividendo, ingresar_divisor)

```

3. Cree una función de interés simple, entregando una excepción si la
tasa es superior a 100.


*   El programa debe recibir: monto, año y tasa.
*   El cálculo de interés será: (monto * año * tasa) / 100
*   Agregue mensajes adecuados al contexto, para la interacción
del programa.







```python
#Método sacar interés, si la tasa es menor a 100 realiza el cálculo
#si la tasa es mayor a 100 envía mensaje de que ingrese un nro menor y no realiza el cálculo
def interes_simple(monto, anio, tasa):
  try:
    if tasa > 100:
      raise ValueError(tasa)
  except ValueError:
    print(f"La tasa de {tasa} es superior a 100, ingresa un nro menor")
  else:
    resultado = (monto * anio * tasa) / 100
    print(resultado)
    #print(tasa, "La tasa es menor a 100")

ingresar_monto = int(input("Ingrese el monto "))
ingresar_anio = int(input("Ingrese los años "))
ingresar_tasa = int(input("Ingrese la tasa "))
resultado = interes_simple(ingresar_monto, ingresar_anio, ingresar_tasa)
```
