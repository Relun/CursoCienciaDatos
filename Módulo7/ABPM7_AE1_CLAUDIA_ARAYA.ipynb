**1.** Defina lo siguiente:


### **a. Inteligencia Artificial**

La inteligencia artificial es una rama de la informática que tiene como objetivo diseñar tecnología que emule la inteligencia humana. Mediante algoritmos, las máquinas pueden llevar a cabo procesos propios de la inteligencia humana, como aprender, razonar o autocorregirse, la IA puede aplicar ese conocimiento para resolver nuevos problemas de manera similar a como lo haría una persona. Mediante la IA se busca contribuir al desarrollo para el ser humano, por ejemplo, la tecnología de IA puede responder de manera significativa a las conversaciones humanas, crear imágenes y textos originales y tomar decisiones basadas en la entrada de datos en tiempo real.

### **b. Machine learning**

Machine learning o aprendizaje de máquinas es una rama de la inteligencia artificial que desarrolla algoritmos que permiten a las máquinas aprender de los datos y mejorar con la experiencia, para así mejorar su funcionamiento. Permite a las máquinas resolver algo para lo cual no está programada, Existen 2 tipos de aprendizaje: Supervisadoo y No Supervisado, el aprendizaje supervisado se utiliza cuando se conocen las etiquetas de los datos, como en la clasificación o la regresión, mientras que el aprendizaje no supervisado se utiliza para explorar y descubrir patrones en los datos sin tener etiquetas

### **c. Deep learning**

El aprendizaje profundo (deep learning) intenta imitar el cerebro
humano, esto hace que el proceso de aprendizaje sea mucho más capaz que el de los módelos estandar de Machine Learning. Deep learning es un subconjunto del aprendizaje automático que consta de una red neuronal con tres o más capas. Si bien una red
neuronal con una sola capa aún puede hacer predicciones aproximadas,
las capas ocultas adicionales pueden ayudar a optimizar y refinar la
precisión.

2. Explique en no más de 5 líneas el concepto de red neuronal para alguien
que no usa ciencia de datos en su vida, dicho de otro modo, explique el
concepto anterior en términos simples.


Una red neuronal es como el cerebro humano, pero en vez de tener neuronas, tiene "nodos", un grupo de nodos forman una "capa" y a su vez el sistema se compone de varias capas, estos nodos se conectan entre sí y al igual que las neuronas del cerebro, se envían datos o información entre ellos, pasando estos datos de capa en capa a través de los nodos que se van activando, al igual como lo hacen las neuronas usando el impulso eléctrico que permiten la comunicación entre neuronas.
