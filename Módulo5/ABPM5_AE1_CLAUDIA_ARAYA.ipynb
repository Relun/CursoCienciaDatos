**Aprendizaje supervisado**


El aprendizaje supervisado es un tipo de aprendizaje automático y es cuando el modelo aprende de una base de datos y transforma un dato en otro a base a los patrones que detecte, por ejemplo se le entregan imágenes etiquetadas con cierta información, entrenamos con esos datos al modelo, luego el modelo es capaz de encontrar esos patrones y transforma esas imagenes a la etiqueta que le hemos entregado.


**Aprendizaje sin supervisión**

Es cuando aprende de una base de datos a agrupar los datos por patrones comunes entre ellos y encontrar esa relación oculta, por ejemplo le entregamos imágenes y cada imagen tiene datos asociados a ella, entrenamos al modelo para que él mismo genere las categorías, entonces agrupará las imágenes según datos que tengan en común entre ellas, creando categorías.

**Aprendizaje por refuerzo**

Es cuando aprende en función a su comportamiento, ya sea por refuerzo positivo o negativo. Dejamos que el modelo haga lo que mejor considere y si ocurren acciones positivas le informamos que lo ha hecho bien a través de recompensas, por ejemplo dándole puntos, si por el contrario, ocurren acciones negativas, le castigamos quitándole puntos.


**Algoritmos de regresión**

En las tareas de regresión, el programa de aprendizaje automático debe estimar y comprender las relaciones entre las variables. El análisis de regresión se enfoca en una variable dependiente y una serie de otras variables cambiantes, lo que lo hace particularmente útil para la predicción y el pronóstico. Existen distintas técnicas usadas en los algoritmos de regresión, por ejemplo:

Regresión lineal (utiliza datos continuos).

Regresión logística (utiliza datos discretos).

Árboles de decisión.


**Algoritmos de agrupación**

Se utilizan en el aprendizaje no supervisado, y sirven para categorizar datos no etiquetados, es decir, datos sin categorías o grupos definidos. El algoritmo funciona mediante la búsqueda de grupos dentro de los datos, con el número de grupos representados por la variable K. A continuación, funciona de manera iterativa para asignar cada punto de datos a uno de los K grupos según las características proporcionadas. Algunos ejemplos:

K-Means Clustering

K-Nearest Neighbours

Cluster por jerarquías



**Algoritmos de Aprendizaje Profundo**

Son un conjunto de algoritmos de aprendizaje automático que están basados en modelos de redes neuronales artificiales con múltiples capas de procesamiento. Estas redes se denominan "profundas" porque tienen muchas capas ocultas entre la entrada y la salida, lo que les permite aprender representaciones de datos cada vez más abstractas y complejas. Algunos ejemplos son:

Redes Neuronales Convolucionales (CNN)

Redes Neuronales Recurrentes (RNN)

Redes Neuronales Generativas Adversarias (GAN)
